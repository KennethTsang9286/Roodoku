```python
import pyautogui
from PIL import ImageGrab
from PIL import ImageOps
from numpy import *
from PIL import Image, ImageEnhance
import pytesseract
import time
```

```python
def cross(A, B):
    "Cross product of elements in A and elements in B."
    return [a+b for a in A for b in B]

digits   = '123456789'
rows     = 'ABCDEFGHI'
cols     = digits
squares  = cross(rows, cols)
unitlist = ([cross(rows, c) for c in cols] +
            [cross(r, cols) for r in rows] +
            [cross(rs, cs) for rs in ('ABC','DEF','GHI') for cs in ('123','456','789')])
units = dict((s, [u for u in unitlist if s in u]) for s in squares)
peers = {}
for s in squares:
    tmp = set()
    for i in range(3):
        tmp = set().union(tmp, set(units[s][i]))
    tmp = tmp - set([s])
    peers[s] = set(tmp)
```

def test():
    assert len(squares) == 81
    assert len(unitlist) == 27
    assert all(len(units[s]) == 3 for s in squares)
    assert all(len(peers[s]) == 20 for s in squares)
    assert units['C2'] == [['A2', 'B2', 'C2', 'D2', 'E2', 'F2', 'G2', 'H2', 'I2'],
                           ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9'],
                           ['A1', 'A2', 'A3', 'B1', 'B2', 'B3', 'C1', 'C2', 'C3']]
    assert peers['C2'] == set(['A2', 'B2', 'D2', 'E2', 'F2', 'G2', 'H2', 'I2',
                               'C1', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9',
                               'A1', 'A3', 'B1', 'B3'])
    print 'All tests pass.'

```python
def parse_grid(grid):
    values = dict((s, digits) for s in squares)
    for s,d in grid_values(grid).items():
        if (d in digits) and (assign(values, s, d) is not False):
            values = assign(values, s, d)
    return values

def grid_values(grid):
    chars = [c for c in grid if c in digits or c in '0.']
    assert len(chars) == 81
    return dict(zip(squares, chars))
```

```python
#s = square, d = digit
def assign(values, s, d):
    if values is False: return False
    other_values = values[s].replace(d, '')
    
    for d2 in other_values:
        values = eliminate(values, s, d2)
    
    if values is False: return False
        
    for ss in peers[s]:
        if (d in values[ss]):
            if eliminate(values, ss, d):
                values = eliminate(values, ss, d)
            else:
                return False
            
    
    assert values != False
    return values


def eliminate(values, s, d):
    if values is False:
        return False
    if d not in values[s]:
        return values 
    values[s] = values[s].replace(d,'')
    
    if len(values[s]) == 0:
        return False 
    elif len(values[s]) == 1:
        d2 = values[s]
        for s2 in peers[s]:
            values = eliminate(values, s2, d2)
    
    if values == False:
        return False
    
    for u in units[s]:
        dplaces = [s for s in u if d in values[s]]
        if len(dplaces) == 0:
            return False 
        elif len(dplaces) == 1:
            if not assign(values, dplaces[0], d):
                return False
            else:
                values = assign(values, dplaces[0],d)
                
    return values
```

```python
def search(values):
    if values is False:
        return False 
    
    tmpp = [len(values[s]) == 1 for s in squares]
    if all(tmpp): 
        return values 
    
    tmp = [None, 10]
    for s in squares:
        if len(values[s]) > 1:
            if tmp[1] > len(values[s]):
                tmp[0] = s
                tmp[1] = len(s)

    tmppp = []
    for d in values[tmp[0]]:
        tmppp.append(search(assign(values.copy(), tmp[0], d)))
    #tmppp =  [search(assign(values.copy(), tmp[0], d)) for d in values[tmp[0]]]
    return some(tmppp)

def some(seq):
    for e in seq:
        if e: return e
    return False

def solve(grid): return search(parse_grid(grid))
```

```python
def getQuestion(xpad, ypad ):
    xlength = 710
    ylength = 710
    box = (xpad, ypad, xpad + xlength, ypad + ylength)
    im = ImageOps.grayscale(ImageGrab.grab(box))
    
    slength = 63
    c = [] 
    ccc = []

    for j in range(9):
        for i in range(9):
            x =  2 + i*slength  + i*17
            y =  j*slength  + j*17
            area = (x,y,x+slength, y+slength)
            newim = im.crop(area) 
            ccc.append(newim.getcolors())
            c.append(newim)
            
    #get white 
    arrayofwhite = []
    arrayofblack = []
    for i in range(len(ccc)):
        if ccc[i][0][1] > 220:
            arrayofwhite.append(i)
        else:
            arrayofblack.append(i)
            
    #image to string of Black
    intensity = 2
    k = []
    for i in arrayofblack:
        xx = c[i]
        xx = xx.convert('1', dither=Image.NONE)
        xx = ImageEnhance.Contrast(xx)
        xx = xx.enhance(intensity)
        k.append(pytesseract.image_to_string(xx, config='-psm 6'))
        
    #translate black
    kk = []
    for x in k:
        tmp = x
        if tmp == 'I': tmp = '1'
        if tmp == 'll': tmp = '4'
        if tmp == '': tmp = '8'
        kk.append(tmp)
        
    #get translated problem
    q = ""
    for i in range(81):
        if i in arrayofwhite:
            q += "."
        else:
            q += kk[arrayofblack.index(i)]
    
    return [q,arrayofwhite]
```
